```python
import numpy as np
import matplotlib.pyplot as plt

flips = np.random.choice(['H', 'T'], size=10000)
heads_ratio = np.cumsum(flips == 'H') / np.arange(1, 10001)

plt.plot(heads_ratio)
plt.axhline(0.5, color='red', linestyle='--')
plt.title('Law of Large Numbers in action')
plt.xlabel('Number of flips')
plt.ylabel('Proportion of heads')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulating the Central Limit Theorem
# Take samples from an Exponential distribution (skewed, not normal)
# Watch the sample MEANS become normal

sample_means = []
for _ in range(5000):
    sample = np.random.exponential(scale=1.0, size=30)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=50, edgecolor='black')
plt.title('Central Limit Theorem — sample means become normal')
plt.xlabel('Sample mean')
plt.ylabel('Frequency')
plt.show()
```

## What I learned today

**Law of Large Numbers:**
As we flip a coin more and more times, the proportion of heads stops 
being random and settles towards the true probability of 0.5. 
More trials = more stability.

**Central Limit Theorem:**
Even if the original data is skewed or messy, if you take the averages 
of many samples, those averages will always form a bell curve. 
This is why normal distribution assumptions work in finance — 
we're usually dealing with averages, not individual values.
